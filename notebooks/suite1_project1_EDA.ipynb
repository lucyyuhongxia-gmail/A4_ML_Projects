# Project 1: Exploratory Data Analysis & ML Types

**Dataset:** California Housing (1990 US Census, 20,640 records)

**Goal:** Learn the three types of Machine Learning through real data exploration. Understand how to load, inspect, and visualize a real dataset.

```
# ====== Setup ======
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
import warnings; warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.dpi'] = 120
DATA_URL = 'https://raw.githubusercontent.com/lucyyuhongxia-gmail/A4_ML_Projects/main/datasets'
def load_data(fn):
    df = pd.read_csv(f'{DATA_URL}/{fn}')
    print(f'Loaded: {fn} -> {df.shape[0]:,} rows x {df.shape[1]} cols')
    return df
print('Setup OK!')
```

## 1. Load the Real Dataset

```
# Load California Housing data from GitHub
# 加州房价数据 (1990年美国人口普查) — 真实数据
df = load_data('housing.csv')

print('=' * 60)
print('DATASET OVERVIEW')
print('=' * 60)
print(f'Rows:    {df.shape[0]:,}')
print(f'Columns: {df.shape[1]}')
print(f'\nColumn names:')
print(list(df.columns))

# Preview first 5 rows — 前5行预览
print('\nFirst 5 rows:')
print(df.head().to_string())

# Data types — 各列数据类型
print('\nData types:')
print(df.dtypes.to_string())
```

## 2. Descriptive Statistics & Quality Check

```
# describe() computes: count, mean, std, min, 25%/50%/75%/max
# 描述性统计 — 快速了解数据分布和范围
print('Numerical column statistics:')
print(df.describe().round(3).to_string())

# Check for missing values — 检查缺失值
print('\nMissing values per column:')
print(df.isnull().sum().to_string())

# Target variable stats — 目标变量 (房价) 统计
print(f'\n=== House Value (MedHouseVal) ==')
print(f'  Mean:   ${df.MedHouseVal.mean()*100:.0f}K')
print(f'  Median: ${df.MedHouseVal.median()*100:.0f}K')
print(f'  Min:    ${df.MedHouseVal.min()*100:.0f}K')
print(f'  Max:    ${df.MedHouseVal.max()*100:.0f}K')
```

## 3. Visual Analysis — Three Key Charts

```
# Chart 1: Target distribution — 房价分布直方图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['MedHouseVal'], bins=50, kde=True, ax=axes[0], color='#2e86de')
axes[0].set_title('House Value Distribution'); axes[0].set_xlabel('Price ($100K)')
sns.boxplot(y=df['MedHouseVal'], ax=axes[1], color='#2e86de')
axes[1].set_title('Box Plot'); plt.tight_layout(); plt.show()
print('Observation: right-skewed (most houses $100-300K, long tail of expensive)')
```

```
# Chart 2: Correlation heatmap — 特征相关性热力图
fig, ax = plt.subplots(figsize=(10, 8))
corr = df.select_dtypes(include=[np.number]).corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, linewidths=1, ax=ax)
ax.set_title('Feature Correlation Matrix', fontsize=14)
plt.tight_layout(); plt.show()

# Find strongest feature for predicting house value
top = corr['MedHouseVal'].drop('MedHouseVal').abs().sort_values(ascending=False)
print('Top features correlated with house value:')
print(top.to_string())
print('\nKey insight: MedInc (median income) has the strongest correlation (r~0.69)')
```

```
# Chart 3: Geographic distribution — 加州房价地图
fig, ax = plt.subplots(figsize=(12, 8))
sc = ax.scatter(df['Longitude'], df['Latitude'],
                c=df['MedHouseVal'], cmap='viridis', alpha=0.3, s=5)
plt.colorbar(sc, ax=ax, label='House Value ($100K)')
ax.set_title('California Housing Prices — Geographic Distribution', fontsize=14)
ax.set_xlabel('Longitude'); ax.set_ylabel('Latitude'); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
print('Observation: coastal areas (LA, SF) are more expensive than inland')
```
